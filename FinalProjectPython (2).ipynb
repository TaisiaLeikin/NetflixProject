```python
def set_primary_key(engine, table_name, column_name):
    with engine.connect() as con:
        con.execute('ALTER TABLE ' + table_name + ' ALTER COLUMN ' + column_name + ' int NOT NULL;')
        con.execute('ALTER TABLE ' + table_name + ' ADD PRIMARY KEY ("' + column_name + '");')

# regions table
set_primary_key(engine , 'Regions', 'RegionID')

# Competition table
set_primary_key(engine , 'Competition', 'ID')

# Competitors table
set_primary_key(engine , 'Competitors', 'CompetitorID')

# GlobalStreamingRevenue table
set_primary_key(engine , 'GlobalStreamingRevenue', 'ID')

#  NetflixRevenueSubscribers table
set_primary_key(engine, 'NetflixRevenueSubscribers', 'ID')
```
